# make model to transcribe the audio

```python
model = WhisperModel("tiny.en", device="cuda" if torch.cuda.is_available() else "cpu" ,  compute_type="int8")
```

```python
source_path = "Test_Data/"
dest_path = "Test_Vtt/"
def format_time(seconds):
    hours = int(seconds // 3600)
    minutes = int((seconds % 3600) // 60)
    seconds = seconds % 60
    return (f"{hours:02}:{minutes:02}:{seconds:06.3f}").replace(".", ",")
def get_vtt(source_path,filename,model):
    
    # Transcribe the audio file with timestamps
    segments, _ = model.transcribe(source_path+filename)
    for segment in segments:
        return [segment.start, segment.end]
timing_segment=[]
for filename in test["audio"]:
    timing_segment.append(get_vtt(source_path,filename,model))
print(timing_segment)
test[["start","end"]] = timing_segment
```
